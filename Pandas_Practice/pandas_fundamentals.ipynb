## Import Library

```python
import pandas as pd
```

## Pandas Series Examples

```python
# Checking pandas version
print(pd.__version__)

# Creating a Series from a list
a = [10, 20, 30]
s = pd.Series(a)
print(s)

# Series with custom index
s = pd.Series(a, index=['x', 'y', 'z'])
print(s)
print(s['y'])

# Series from dictionary
d = {"d1":200, "d2":300, "d3":400}
s = pd.Series(d)
print(s)

# Selecting specific elements
s = pd.Series(d, index=["d1", "d2"])
print(s)
```

## DataFrame Basics

```python
# Creating DataFrame from dictionary

data = {"a":[10,20,30], "b":[40,50,60]}
df = pd.DataFrame(data)

print(df)

# Accessing rows using loc
print(df.loc[0])
print(df.loc[0:2])
```

## DataFrame Indexing

```python
# DataFrame with custom index

data = {"a":[10,20,30], "b":[40,50,60]}
df = pd.DataFrame(data, index=["x","y","z"])

print(df)

# Access specific row
print(df.loc['y'])

# Access specific element
print(df.loc['y']['b'])

# Access column
print(df['b'])

# Slicing rows
print(df.loc[::2]["b"])
```

## Load CSV Dataset

```python
# Loading dataset
df = pd.read_csv("students.csv")

# Inspect dataset
print(df.head())
print(df.info())
print(df.shape)
```

## Checking Missing Values

```python
print(df.isnull())
print(df.isnull().sum())
```

## Handling Missing Values

```python
# Fill missing values using mean
df["Maths"] = df["Maths"].fillna(df["Maths"].mean())
df["Science"] = df["Science"].fillna(df["Science"].mean())
df["English"] = df["English"].fillna(df["English"].mean())
```

## Feature Engineering

```python
# Adding Total and Average columns

df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3

print(df.head())
```

## Pass/Fail Classification

```python
# Creating Result column

df["Result"] = df["Average"].apply(lambda x: "Pass" if x >= 50 else "Fail")

print(df.head())
```

## Filtering Data

```python
# Students with attendance > 90
print(df[df["Attendance"] > 90]["Name"])

# Students who passed
print(df[df["Result"] == "Pass"]["Student_ID"])

# Students whose Maths marks > English
print(df[df["Maths"] > df["English"]]["Student_ID"])
```

## String Filtering

```python
# Students whose name ends with 'a'
print(df[df["Name"].str[-1] == "a"]["Name"])
```
