#  KNN Classification
This notebook demonstrates K-Nearest Neighbors classification on the sklearn digits dataset.

## 1.Imports

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from collections import Counter
```

## 2.Load Data

```python
digits = load_digits()
X, y = digits.data, digits.target
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

## 3.KNN Function

```python
def knn_predict(X_train, y_train, X_test, k):
    y_pred = []
    for test_point in X_test:
        distances = np.sqrt(np.sum((X_train - test_point) ** 2, axis=1))
        nearest_indices = distances.argsort()[:k]
        nearest_labels = y_train[nearest_indices]
        most_common = Counter(nearest_labels).most_common(1)[0][0]
        y_pred.append(most_common)
    return np.array(y_pred)
```

## 4.Evaluate Accuracy and Find Optimal K

```python
k_values = range(1, 21)
acc_list = []

print("K\tAccuracy\tMisclassified")
print("-" * 35)
for k in k_values:
    y_pred = knn_predict(X_train, y_train, X_test, k)
    misclassified = np.sum(y_pred != y_test)
    acc = np.mean(y_pred == y_test)
    acc_list.append(acc)
    print(f"{k}\t{acc:.4f}\t\t{misclassified}")

optimal_k = k_values[np.argmax(acc_list)]
best_acc = max(acc_list)

print("\nMost optimal K:", optimal_k)
print(f"Best Accuracy: {best_acc:.4f}")
```
